## Impact Number of Attendance have on the Performance against A&E 4-hour Standard

**Insight**

DESCRIPTION - 
There is a negative correlation between percentage within 4-hours and attendance. This means as attendance increases, performance against the 4-hour A&E decreases. Therefore, increase in attendance has a negative impact on performance against A&E 4 hour a standard. 


```python
rg = sns.regplot(x = 'attendance', y = 'percentage_within_4_hours', data = phs1, ci = None, line_kws = {'color':'red'})
rg.set_title('Impact of attendance on the Performance against the A&E Standard')
```

## Performance across NHS Scotland Health Boards

**Insight**

DESCRIPTION - NHS Orkney, NHS Shetland and NHS Western perform well with percentage within 4 hours above 95%. 
NHS forth valley, NHS Lanarkshire and NHS Borders were trailing at the bottom with performance below 80%.

```python
bx1  = sns.boxplot(x = 'nhs_board_name', y = 'percentage_within_4_hours', data = phs1, palette = 'coolwarm')
bx1.set_xticklabels(bx1.get_xticklabels(), rotation = 90);
bx1.set_title('Performance against the A&E standard acoss NHS Scotland Health Boards');
```

## Performance across NHS Hospitals

**Insight**

DESCRIPTION - Gilbert Bain Hospital, Royal Aberdeen Children Hospital performed well with percentage_within_4_hours above 95%.
Royal Infirmary of Edinburgh, University Hospital Wishaw were trailing behind with performance below 70%.

```python
bx2  = sns.boxplot(y = 'percentage_within_4_hours', x = 'location_name', data = phs1, palette = 'coolwarm')
bx2.set_xticklabels(bx2.get_xticklabels(), rotation = 90);
bx2.set_title('Performance against the A&E standard acoss NHS Scotland Hospitals')
```

## Lineplot of Performance against the A&E Standard for the year 2021

**Insight**

DESCRIPTION - The spikes in the data shows fluctuations in the weekly data for percentage within 4 hours.
Performance against the A&E 4-hour standard increased in winter 2020/21 but started dropping during spring 2021.
The drop continued through summer 2021 and autumn 2021. By autumn 2021 performance against the A&E 4-hour standard was below 80% remained at this rate for a prolonged period of time.

```python
sns.set(rc={'figure.figsize':(15,9)})
Ln1 = sns.lineplot(x = 'week_ending_date', y = 'percentage_within_4_hours', ci  = None, marker = 0, data = phs1)
sns.set_style('white')
Ln1.axhline(82, ls='--', color = 'green')
Ln1.set_xticklabels(Ln1.get_xticklabels(), rotation = 90);
Ln1.set_title('Performance across NHS Scotland throughout the year 2021')
Ln1.set_xticks(range(len(phs1.week_ending_date.unique()))) # <--- set the ticks first
Ln1.set_xticklabels(phs1.week_ending_date.unique());
```

## Lineplot for attendance across NHS Scotland in 2021

**Insight**

DESCRIPTION - There was a decrease in attendance in Winter 2020/21 which lasted up to February 2021 probabably due to COVD-19 restrictions but while performance increased in the same period.
After February 2021, attendance rose sharply in spring 2021 while performance against the A&E 4-hour standard began to fall. 
At the beginning of winter 2021, Total attendance began to fall while performance against the A&E waiting time began to rise.

```python
sns.set(rc={'figure.figsize':(15,9)})
Ln2 = sns.lineplot(x = 'week_ending_date', y = 'attendance', data = phs1, ci = None)
Ln2.set_title('Attendance in A&E in 2021 across NHS Scotland')
Ln2.axhline(800, ls='--', color = 'green')
Ln2.set_xticklabels(Ln2.get_xticklabels(), rotation = 90);
Ln2.set_xticks(range(len(phs1.week_ending_date.unique()))); # <--- set the ticks first
Ln2.set_xticklabels(phs1.week_ending_date.unique());
```
